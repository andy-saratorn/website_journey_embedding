# Model tuning

```python
def prep_cols(_df, df_joe):
    _df = filter_joe_unique_values(_df, df_joe, ft_lists=[['browser'],['os','locale'], ['location']])
    _df = get_localtime(_df)
    _df['hour'] = _df['local_time'].dt.hour
    _df['dow'] = _df['local_time'].dt.day_of_week
    _df['train_test_label'] = np.random.choice(2, _df.shape[0], replace=True, p=[0.8,0.2])
    _df['idx'] = range(_df.shape[0])
    if 'user_id' not in _df.columns:
        _df['user_id'] = -99
    return _df
```

```python
def prep_df_like_joe(_df):
    cond1 = _df['gender'] == 'm'
    cond2 = (_df['browser'] != 'others') & (_df['locale'] != 'others')
    cond3 = (_df['location'] != 'UTC')
    cond4 = _df['hour'].isin([15,  4, 14, 13,  5,  6])
    cond5 = _df['top_site_first'] = 1
    _df = _df[cond1 & cond2 & cond3 & cond4]
    return _df
```

## Sites

```python
def get_df_site_expld(_df):
    df_site = _df.copy()
    df_site = df_site[['idx','user_id','local_time','hour','sites']]
    df_site = df_site.explode('sites')
    df_site['row'] = range(df_site.shape[0])
    df_site['site'] = df_site['sites'].apply(lambda x: x['site'])
    df_site['length'] = df_site['sites'].apply(lambda x: x['length'])
    return df_site
```

```python
df_train_sites = get_df_site_expld(df_train)
```

```python
sites = df_train_sites[['site']].drop_duplicates()
sites['site_id'] = np.arange(sites.shape[0]) + 1
```

```python
df_site_joe = df_train_sites[df_train_sites['user_id'] == 0].copy()
df_site_joe = df_site_joe.groupby('site')['length'].sum().reset_index()
df_site_joe['pcnt'] = df_site_joe['length'] / df_site_joe['length'].sum()

df_site_not_joe = df_train_sites[df_train_sites['user_id'] != 0].copy()
df_site_not_joe = df_site_not_joe.groupby('site')['length'].sum().reset_index()
df_site_not_joe['pcnt'] = df_site_not_joe['length'] / df_site_not_joe['length'].sum()

df_site_map = df_site_joe.merge(df_site_not_joe, 
                                on='site',
                                how='outer',
                                suffixes=('','_nj')).fillna(0)
df_site_map['pcnt_diff'] = df_site_map['pcnt'] - df_site_map['pcnt_nj']
df_site_map['pcnt_diff_rank_a'] = df_site_map['pcnt_diff'].rank(ascending=True)
df_site_map['pcnt_diff_rank_d'] = df_site_map['pcnt_diff'].rank(ascending=False)
cond = (df_site_map['pcnt_diff_rank_a'] <= 0) | (df_site_map['pcnt_diff_rank_d'] <=5)
```

```python
top_sites = df_site_map.loc[cond, ['site']]
top_sites = top_sites.merge(sites, on='site')
top_sites
```

```python
def top_sites_ft(df_site, joe_top_sites):
    df_top_sites = df_site.copy()
    df_top_sites['row'] = range(df_top_sites.shape[0])
    df_top_sites['first'] = df_top_sites.groupby('idx')['row'].rank()
    df_top_sites.loc[df_top_sites['first'] > 1, 'first'] = 0
    df_top_sites['last'] = df_top_sites.groupby('idx')['row'].rank(ascending=False)
    df_top_sites.loc[df_top_sites['last'] > 1, 'last'] = 0

    df_top_sites = df_top_sites.groupby(['idx','site'])[['length','first','last']].sum().reset_index()
    df_top_sites['top_sites'] = df_top_sites['site'].isin(joe_top_sites)
    for x in ['length','first','last']:
        df_top_sites[f'top_site_{x}'] = df_top_sites[x]
        df_top_sites.loc[~df_top_sites['top_sites'], f'top_site_{x}'] = 0
        df_top_sites[f'top_site_{x}'] = df_top_sites.groupby('idx')[f'top_site_{x}'].transform('max')
    agg_dict = {'top_sites':"sum", 
                'top_site_length':'min',
                'top_site_first':'max', 
                'top_site_last':'max'}
    df_top_sites = df_top_sites.groupby('idx').agg(agg_dict).reset_index()
    return df_top_sites
```

```python
df_train_top_sites = top_sites_ft(df_train_sites, top_sites['site'])
df_train_map = df_train.merge(df_train_top_sites, on='idx')
df_train_lk_j = prep_df_like_joe(df_train_map)
```

```python
df_test_sites = get_df_site_expld(df_test)
df_test_top_sites = top_sites_ft(df_test_sites, top_sites['site'])
df_test_map = df_test.merge(df_test_top_sites, on='idx')
df_test_lk_j = prep_df_like_joe(df_test_map)
```

```python
def top_sites_length(_df, sites, top_sites):
    _df = get_df_site_expld(_df)
    _df['total_length'] = _df.groupby('idx')['length'].transform('sum')
    _df = _df.merge(sites, on='site')
    _df = _df.merge(top_sites[['site_id']], on='site_id', how='inner')
    _df = _df.groupby(['idx','total_length','site_id'])['length'].sum().reset_index()
    _df = _df.pivot_table(index=['idx','total_length'], columns='site_id', values='length').fillna(0)
    _df.columns = [f's_{x}' for x in _df.columns]
    _df = _df.reset_index()
    return _df
```

## Sites Embedding

### Doc2Vec

```python
X_train_doc = get_df_site_expld(df_train)
X_train_doc = X_train_doc.merge(sites, on='site')
X_train_doc = X_train_doc.sort_values('row')
X_train_doc['rank'] = X_train_doc.groupby('idx')['row'].rank(method='first').astype(int)
max_rank = X_train_doc['rank'].max()
X_train_doc = X_train_doc.pivot_table(index='idx', columns='rank', values='site_id').fillna(0)
X_train_doc = X_train_doc.astype(int).astype(str)
X_train_doc = X_train_doc.apply(lambda x: list(x),1)
tagged_data = [TaggedDocument(d, [i]) for i, d in enumerate(X_train_doc)]
```

```python
doc2vec = Doc2Vec(tagged_data, 
                  vector_size=100,
                  window=2, 
                  min_count=1,
                  seed=20,
                  epochs=20,
                  workers=1)
```

```python
emb_df_doc = pd.DataFrame(
        [doc2vec.dv.get_vector(n) for n in doc2vec.dv.index_to_key]
    )
```

```python
def prep_X_doc2vec(_df, sites, max_rank, doc2vec, pca_doc):

    _df = get_df_site_expld(_df)

    _df = _df.merge(sites, on='site')
    _df['rank'] = _df.groupby('idx')['local_time'].rank(method='first').astype(int)
    max_rank = _df['rank'].max()
    _df = _df.pivot_table(index=['idx','user_id'], columns='rank', values='site_id').fillna(0)

    for x in range(1,max_rank+1):
        if x not in _df.columns:
            _df[x] = 0
            
    X = _df.astype(int).astype(str)
    X = X.apply(lambda x: list(x),1)

    emb_df = []
    for i in range(_df.shape[0]):
        emb_df.append(doc2vec.infer_vector(X.iloc[i], ))
    emb_df = pd.DataFrame(pca_doc.transform(emb_df), index=_df.index)
    return emb_df
```

```python
n_pca = 2
pca_doc = PCA(n_components=n_pca, random_state=20)
pca_doc.fit(emb_df_doc)
```

```python
ros = RandomOverSampler(random_state=20)
```

```python
df_train_d2v = prep_X_doc2vec(df_train_lk_j, sites, max_rank, doc2vec, pca_doc).reset_index()
df_train_d2v['joe'] = df_train_d2v['user_id'] == 0
X_train = df_train_d2v[[x for x in range(n_pca)]]
y_train = df_train_d2v['user_id'] == 0

transformer = Normalizer().fit(X_train)
X_train = transformer.transform(X_train)

# fit predictor and target variable
X_train_ros, y_train_ros = ros.fit_resample(X_train, y_train)
```

```python
df_test_d2v = prep_X_doc2vec(df_test_lk_j, sites, max_rank, doc2vec, pca_doc).reset_index()
df_test_d2v['joe'] = df_test_d2v['user_id'] == 0
X_test = df_test_d2v[[x for x in range(n_pca)]]
y_test = df_test_d2v['user_id'] == 0

X_test = transformer.transform(X_test)
```

## SVM

```python
df_score_SVM = None
for d in notebook.tqdm(range(1,6)):
    cv_scores = []
    kf = KFold(n_splits=5, shuffle=True, random_state=20)
    for train_idx, test_idx in kf.split(X_train_ros):
        clf_fit = svm.SVC(degree=d, probability=True).fit(X_train_ros[train_idx], y_train_ros[train_idx])
        c = roc_auc_score(y_train_ros[test_idx], clf_fit.predict_proba(X_train_ros[test_idx])[:, 1])
        cv_scores.append(c)
    df_score_i = pd.DataFrame(cv_scores, columns=['roc_auc'])
    df_score_i['degree'] = d
    df_score_i['gini'] = 2*df_score_i['roc_auc'] - 1

    df_score_SVM = pd.concat([df_score_SVM, df_score_i])
```

```python
figure = plt.figure(facecolor='white')
sns.lineplot(x='degree', y='gini', data=df_score_SVM)
opt_degree = 1
plt.axvline(opt_degree)
plt.title('SVM')
plt.show()
```

## Best model

```python
df_roc_SVM = None
i = 0
for proba in notebook.tqdm(np.arange(0.1,0.51,0.02)):
    kf = KFold(n_splits=5, shuffle=True, random_state=2)
    for train_idx, test_idx in kf.split(X_train_ros):
        clf_fit = svm.SVC(degree=opt_degree, probability=True).fit(X_train_ros[train_idx], y_train_ros[train_idx])
        y_test_predict = clf_fit.predict_proba(X_train_ros[test_idx])[:,1] > proba
        cfs_mtrx = confusion_matrix(y_train_ros[test_idx] ,y_test_predict)
        tn, fp, fn, tp = cfs_mtrx.ravel()
        fpr = fp / (fp + tn)
        tpr = tp / (tp + fn)
        df_roc_i = pd.DataFrame({'cv':[i]})
        df_roc_i['cv'] = i
        df_roc_i['fpr'] = fpr
        df_roc_i['tpr'] = tpr
        df_roc_i['proba'] = proba
        df_roc_SVM = pd.concat([df_roc_SVM, df_roc_i])
        i += 1
```

```python
figure = plt.figure(facecolor='white')
_df = df_roc_SVM.melt(id_vars='proba', value_vars=['fpr','tpr'])
sns.lineplot(x='proba', y='value', hue='variable', data=_df)

thres = 0.34
plt.axvline(x=thres)
plt.title('SVM')
plt.show()
```
