```python
_df_unique_vals = df_train.copy()
_df_unique_vals['joe'] = _df_unique_vals['user_id'] == 0
_df_unique_vals['row'] = 1
for x in ['browser','os','locale', 'gender', 'location']:
    if isinstance(x, list):
        tmp = _df_unique_vals.groupby(['joe',*x])['row'].sum().reset_index()
    else:
        tmp = _df_unique_vals.groupby(['joe',x])['row'].sum().reset_index()
    
    tmp['pcnt'] = tmp['row'] / tmp.groupby('joe')['row'].transform('sum')
    figure = plt.figure(facecolor='white')
    sns.barplot(x='pcnt', y=x, hue='joe', data=tmp)
    plt.title(x)
    plt.show()
```

# Sites

## Top sites

```python
df_site = df_filter.copy()
df_site = df_site[['idx','user_id','local_time','sites']]
df_site = df_site.explode('sites')
df_site['site'] = df_site['sites'].apply(lambda x: x['site'])
df_site['length'] = df_site['sites'].apply(lambda x: x['length'])

sites = df_site[['site']].drop_duplicates()
sites['site_id'] = np.arange(sites.shape[0]) + 1

df_site = df_site.merge(sites, on='site')
```

```python
df_site_joe = df_site[df_site['user_id'] == 0].copy()
df_site_joe_agg = df_site_joe.groupby('site')['length'].sum().reset_index()
df_site_joe_agg['pcnt'] = df_site_joe_agg['length'] / df_site_joe_agg['length'].sum()
df_site_joe_agg = df_site_joe_agg.sort_values('length',ascending=False)
df_site_joe_agg['cum_pcnt'] = df_site_joe_agg['pcnt'].cumsum()
```

```python
df_site_not_joe = df_site[df_site['user_id'] != 0].copy()
df_site_not_joe = df_site_not_joe.groupby('site')['length'].sum().reset_index()
df_site_not_joe['pcnt'] = df_site_not_joe['length'] / df_site_not_joe['length'].sum()
```

```python
df_site_map = df_site_joe_agg.merge(df_site_not_joe, 
                                on='site',
                                how='outer',
                                suffixes=('','_nj')).fillna(0)
df_site_map['pcnt_diff'] = df_site_map['pcnt'] - df_site_map['pcnt_nj']
df_site_map['rank_d'] = df_site_map['pcnt_diff'].rank(ascending=False)
df_site_map['rank_a'] = df_site_map['pcnt_diff'].rank(ascending=True)
cond1 = (df_site_map['rank_d'] <= 5)
cond2 = (df_site_map['rank_a'] <= 5)
df_site_map = df_site_map[cond1 | cond2]
```

```python
_df = df_site_map.melt(id_vars='site', value_vars=['pcnt','pcnt_nj'])
figure = plt.figure(facecolor='white', figsize=(15,4))
sns.barplot(x='site', 
            y='value', 
            hue='variable',
            palette=('tab:orange', 'tab:blue'),
            data=_df)
```

```python
joe_top_sites = df_site_map.loc[cond1, 'site']
```

```python
df_joe_top_sites = df_site_joe.copy()
df_joe_top_sites['row'] = range(df_joe_top_sites.shape[0])
df_joe_top_sites['first'] = df_joe_top_sites.groupby('idx')['row'].rank()
df_joe_top_sites.loc[df_joe_top_sites['first'] > 1, 'first'] = 0
df_joe_top_sites['last'] = df_joe_top_sites.groupby('idx')['row'].rank(ascending=False)
df_joe_top_sites.loc[df_joe_top_sites['last'] > 1, 'last'] = 0

df_joe_top_sites = df_joe_top_sites.groupby(['idx','site'])[['length','first','last']].sum().reset_index()
df_joe_top_sites['top_sites'] = df_joe_top_sites['site'].isin(joe_top_sites)
for x in ['length','first','last']:
    df_joe_top_sites[f'top_site_{x}'] = df_joe_top_sites[x]
    df_joe_top_sites.loc[~df_joe_top_sites['top_sites'], f'top_site_{x}'] = 0
    df_joe_top_sites[f'top_site_{x}'] = df_joe_top_sites.groupby('idx')[f'top_site_{x}'].transform('max')
agg_dict = {'top_sites':"sum", 
            'top_site_length':'min',
            'top_site_first':'max', 
            'top_site_last':'max'}
df_joe_top_sites = df_joe_top_sites.groupby('idx').agg(agg_dict).reset_index()
```

## Embedding

```python
X = df_site.copy()
X['rank'] = X.groupby('idx')['local_time'].rank(method='first')
X = X.pivot_table(index='idx', columns='rank', values='site_id').fillna(0)
X = X.astype(int)
X = X.apply(lambda x: list(x),1)
X.shape
```

```python
# train word2vec model
model = Word2Vec(window=10, sg=1, min_count=1, 
                 alpha=0.03, min_alpha=0.0007,
                 seed=20)

model.build_vocab(X, progress_per=20)

model.train(X, total_examples = model.corpus_count, 
            epochs=10, report_delay=1)
```

```python
print(model)
```

```python
emb_df = pd.DataFrame(
        [model.wv.get_vector(n) for n in model.wv.key_to_index],
        index = model.wv.key_to_index.keys()
    )
```

```python
n_pca = 6
pca = PCA(n_components=n_pca, random_state=20)
pca.fit(emb_df)
print(pca.explained_variance_ratio_)
pca_mdl = pca.transform(emb_df)

emb_df_PCA = pd.DataFrame(
        pca_mdl,
        columns=[f'pca{x+1}' for x in range(n_pca)]
    )
emb_df_PCA['site_id'] = emb_df.index
```

```python
df_site_length = df_site.copy()
df_site_length['row'] = range(df_site_length.shape[0])
df_site_length['first'] = df_site_length.groupby('idx')['row'].rank()
df_site_length.loc[df_site_length['first'] > 1, 'first'] = 0
df_site_length['last'] = df_site_length.groupby('idx')['row'].rank(ascending=False)
df_site_length.loc[df_site_length['last'] > 1, 'last'] = 0

df_site_length = df_site_length.groupby(['idx','site'])[['length','first','last']].sum().reset_index()
df_site_length['top_sites'] = df_site_length['site'].isin(joe_top_sites)
for x in ['length','first','last']:
    df_site_length[f'top_site_{x}'] = df_site_length[x]
    df_site_length.loc[~df_site_length['top_sites'], f'top_site_{x}'] = 0
    df_site_length[f'top_site_{x}'] = df_site_length.groupby('idx')[f'top_site_{x}'].transform('max')
agg_dict = {'top_sites':"sum", 
            'top_site_length':'min',
            'top_site_first':'max', 
            'top_site_last':'max'}
df_site_length = df_site_length.groupby('idx').agg(agg_dict).reset_index()
```

```python
df_lk_joe = df_filter.copy()
cond1 = df_lk_joe['gender'] == 'm'
cond2 = df_lk_joe['browser'] != 'others'
cond3 = df_lk_joe['os'] != 'others'
cond4 = df_lk_joe['locale'] != 'others'
cond5 = df_lk_joe['location'] != 'UTC'
cond6 = df_lk_joe['hour'].isin([4,5,6,13,14,15])
df_lk_joe = df_lk_joe[cond1&cond2&cond3&cond4&cond5&cond6]
df_lk_joe = df_lk_joe.merge(df_site_length, on='idx')
df_lk_joe = df_lk_joe[df_lk_joe['top_site_length'].between(100,900)]
df_lk_joe = df_lk_joe[df_lk_joe['top_sites'] > 1]
df_lk_joe = df_lk_joe[df_lk_joe['top_site_first'] == 1]
df_lk_joe = df_lk_joe[df_lk_joe['top_site_last'] == 0]
df_lk_joe['joe'] = df_lk_joe['user_id'] == 0
df_lk_joe.shape
```

```python
df_site_lk_joe = df_lk_joe.copy()
df_site_lk_joe = df_site_lk_joe[['idx','user_id','local_time','sites']]
df_site_lk_joe = df_site_lk_joe.explode('sites')
df_site_lk_joe['site'] = df_site_lk_joe['sites'].apply(lambda x: x['site'])
df_site_lk_joe['length'] = df_site_lk_joe['sites'].apply(lambda x: x['length'])
df_site_lk_joe = df_site_lk_joe.merge(sites, on='site')
```

```python
df_site_pca = df_site_lk_joe.merge(emb_df_PCA, on='site_id', how='left')
df_site_pca = df_site_pca[df_site_pca['site'].isin(joe_top_sites)]
df_site_pca['joe'] = df_site_pca['user_id'] == 0
df_site_pca['weight'] = df_site_pca['length'] / df_site_pca.groupby('idx')['length'].transform('sum')
for x in range(n_pca):
    df_site_pca[f'w_pca_{x+1}'] = df_site_pca[f'pca{x+1}'] * df_site_pca['weight']

w_pca_cols = [x for x in df_site_pca.columns if 'w_pca' in x]
df_site_pca = df_site_pca.groupby(['idx','user_id','joe'])[w_pca_cols].mean().reset_index()
```
